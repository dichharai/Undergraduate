```python
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
volcano = pd.read_table('volcano.txt')
volcano
```

```python
#removing 0 indexed row
volcano = volcano.drop(volcano.index[[0]])
volcano
```

```python
#removing Number column
volcano = volcano.drop('Number',1)
#volcano = volcano.drop('Country',1)
volcano
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```
